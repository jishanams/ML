```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Titanic.csv')
df
```

titanic

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.drop(['PassengerId','Ticket','Cabin','Name'],axis=1,inplace=True) 
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l_sex=LabelEncoder()
l_emb=LabelEncoder()
df['Sex']=l_sex.fit_transform(df['Sex'])
df['Embarked']=l_emb.fit_transform(df['Embarked'])
```

```python
x=df.drop('Survived',axis=1)
y=df[['Survived']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
import pickle
pickle.dump(model,open('model.pkl','wb'))
pickle.dump(l_sex,open('l_sex.pkl','wb'))
pickle.dump(l_emb,open('l_emb.pkl','wb'))
```

```python
model=pickle.load(open('model.pkl','rb'))
l_sex=pickle.load(open('l_sex.pkl','rb'))
l_emb=pickle.load(open('l_emb.pkl','rb'))
```

```python
#predicition
pclass=int(input('enter the pclass:'))
sex=input('enter the sex:')
age=int(input('enter the age:'))
sibsp=int(input('enter sibsp:'))
parch=int(input('enter parch:'))
fare=float(input('enter fare:'))
embarked=input('enter embarked:')
l_sex=l_sex.transform([sex])[0]
embarked_l=l_emb.transform([embarked])[0]
predict=model.predict([[pclass,l_sex,age,sibsp,parch,fare,embarked_l]])
if predict==1:
    print('Survived')
else:
    print('Did not survive')
```
